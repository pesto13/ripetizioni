Assegna alla lista studenti il valore `["Arianna","Bruno","Dario","Elena","Mario","Teresa"]`

* aggiungi in fondo alla lista il nome tuo e del tuo vicino di banco
* inserisci in posizione di indice 3 il nome di un altro tuo compagno di classe
* rimuovi lo studente in seconda posizione
* cancella la studentessa “Teresa”
* stampa indice e nome di ogni studente della lista


````python
studenti = ["Arianna","Bruno","Dario","Elena","Mario","Teresa"]

studenti.append("Shanza")
studenti.append("Carlo")


studenti.insert(3, "Mario")

del(studenti[2])

studenti.remove("Teresa")


for i in range(len(studenti)):
    print(studenti[i])
````
